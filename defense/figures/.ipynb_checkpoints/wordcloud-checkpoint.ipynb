```python
def generate_wordcloud(
    text,
    list_of_stopwords=[],
):
    return WordCloud(
        background_color="white",
        stopwords=list_of_stopwords,
        max_words=200,
        #min_font_size=10,
        max_font_size=80,
        random_state=42,
    ).generate(text)
```

```python
def color_words(*args, **kwargs):
    import random
    return "rgb({red}, {green}, {blue})".format(
        red=0,
        green=random.randint(50, 150),
        blue=random.randint(100, 250)
    )
```

```python
plt.imshow(
    generate_wordcloud(
        text=text,
        list_of_stopwords=list_of_stopwords,
    ).recolor(color_func=color_words)
)
plt.axis("off")
plt.show()
```
